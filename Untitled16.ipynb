```python
from pydub import AudioSegment
import os

def convert_to_standard_wav(input_path):
    """Converte um arquivo de √°udio para um WAV compat√≠vel"""
    output_path = input_path.replace(".wav", "_converted.wav")

    try:
        audio = AudioSegment.from_file(input_path)
        audio = audio.set_frame_rate(16000).set_channels(1)  # Ajuste para 16kHz mono (compat√≠vel com ASR/NLP)
        audio.export(output_path, format="wav")
        print(f"Arquivo convertido e salvo em: {output_path}")
        return output_path
    except Exception as e:
        print(f"Erro ao converter {input_path}: {e}")
        return None

# Caminho do arquivo original
input_file = "/content/drive/MyDrive/DAIC-WOZ_Data/431_P/431_AUDIO.wav"

# Converter para WAV padr√£o
converted_file = convert_to_standard_wav(input_file)

if converted_file:
    print("Convers√£o conclu√≠da. Agora o arquivo pode ser lido com `librosa` ou `scipy`.")
else:
    print("Falha na convers√£o. Verifique o arquivo original.")
```
